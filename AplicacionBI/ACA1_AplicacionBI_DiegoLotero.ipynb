Definir las librerias

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Importar la data, utilizando dataFrame

```python
dataSet = pd.read_csv("https://raw.githubusercontent.com/Dltr17/AnalisisExportaciones/main/AplicacionBI/ACA1-%20DiegoLotero%20-%20AplicacionBI.csv")
```

Primeras filas de datos

```python
print("Exploracion de Datos:")
dataSet.head()
```

DescripciÃ³n estadÃ­stica

```python
print("\nDescripciÃ³n estadistica:")
dataSet.describe().T
```

Top cinco paÃ­ses con mayor exportaciones

```python
top_paises = dataSet.groupby('COD_PAI4').agg({
    'FOBDOL': 'sum',          # Suma de exportaciones
    'CANTI': 'sum',           # Suma de cantidad de productos exportados
    }).sort_values(by='FOBDOL', ascending=False).head(5)

print("Top 5 paÃ­ses de exportaciÃ³n")
top_paises
```

GrÃ¡fico de barras para visualizar los 5 paÃ­ses principales

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=top_paises.index, y=top_paises['FOBDOL'], palette='viridis')
plt.title('Top 5 PaÃ­ses de ExportaciÃ³n')
plt.xlabel('CÃ³digo del PaÃ­s')
plt.ylabel('Valor de Exportaciones (FOBDOL)')
plt.show()
```

Top 5 Departamentos de ExportaciÃ³n

```python
top_dept = dataSet.groupby( 'COD_SAL').agg({
    'FOBDOL': 'sum',          # Suma de exportaciones
    'CANTI': 'sum',           # Suma de cantidad de productos exportados
}).sort_values(by='FOBDOL', ascending=False).head(5)

print("Top 5 Departamentos de exportaciÃ³n:")
print(top_dept)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_dept.index, y=top_dept['FOBDOL'], palette='viridis')
plt.title('Top 5 Departamentos de ExportaciÃ³n')
plt.xlabel('CÃ³digo de Salida')
plt.ylabel('Valor de Exportaciones (FOBDOL)')
plt.show()
```

GrÃ¡fico de DispersiÃ³n: FOBDOL en funciÃ³n de CANTI Septiembre

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='CANTI', y='FOBDOL', data=dataSet)
plt.title('GrÃ¡fico de DispersiÃ³n: FOBDOL en funciÃ³n de CANTI' )
plt.xlabel('Cantidad de Productos Exportados (CANTI)')
plt.ylabel('FOBDOL (Valor de Exportaciones en DÃ³lares)')
plt.show()
```
